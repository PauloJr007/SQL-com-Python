```python
#Importando as bibliotecas necessárias para criar os dados

import pandas as pd # Biblioteca para manipulação dos dados.
import random # Biblioteca para gerar registros aleatórios.
from faker import Faker # Biblioteca que gera dados ficticios.
```

```python
#Gerando Dados e transformando em DataFrames

fake = Faker()

# Número de registros
num_clientes = 5000
num_faturas = 7000
num_pagamentos = 6000
num_despesas = 4000
num_receitas = 5000

# Gerar Clientes
clientes = []
for i in range(1, num_clientes + 1):
    clientes.append([
        i,
        fake.name(),
        fake.email(),
        fake.phone_number(),
        fake.company()
    ])

df_clientes = pd.DataFrame(clientes, columns=["id_cliente", "nome", "email", "telefone", "empresa"])

# Gerar Faturas
faturas = []
for i in range(1, num_faturas + 1):
    id_cliente = random.randint(1, num_clientes)
    data_emissao = fake.date_between(start_date='-2y', end_date='today')
    data_vencimento = fake.date_between(start_date=data_emissao, end_date='+30d')
    valor_total = round(random.uniform(100, 10000), 2)
    status = random.choice(["Paga", "Pendente", "Atrasada"])
    
    faturas.append([i, id_cliente, data_emissao, data_vencimento, valor_total, status])

df_faturas = pd.DataFrame(faturas, columns=["id_fatura", "id_cliente", "data_emissao", "data_vencimento", "valor_total", "status"])

# Gerar Pagamentos
pagamentos = []
for i in range(1, num_pagamentos + 1):
    id_fatura = random.randint(1, num_faturas)
    data_pagamento = fake.date_between(start_date='-2y', end_date='today')
    valor_pago = round(random.uniform(50, 10000), 2)
    metodo_pagamento = random.choice(["Cartão de Crédito", "Boleto", "PIX", "Transferência"])
    
    pagamentos.append([i, id_fatura, data_pagamento, valor_pago, metodo_pagamento])

df_pagamentos = pd.DataFrame(pagamentos, columns=["id_pagamento", "id_fatura", "data_pagamento", "valor_pago", "metodo_pagamento"])

# Gerar Despesas
despesas = []
categorias_despesas = ["Salários", "Infraestrutura", "Marketing", "TI", "Operacional"]
for i in range(1, num_despesas + 1):
    categoria = random.choice(categorias_despesas)
    descricao = fake.sentence(nb_words=6)
    valor = round(random.uniform(500, 50000), 2)
    data_despesa = fake.date_between(start_date='-2y', end_date='today')
    
    despesas.append([i, categoria, descricao, valor, data_despesa])

df_despesas = pd.DataFrame(despesas, columns=["id_despesa", "categoria", "descricao", "valor", "data_despesa"])

# Gerar Receitas
receitas = []
for i in range(1, num_receitas + 1):
    id_cliente = random.randint(1, num_clientes)
    descricao = fake.sentence(nb_words=6)
    valor = round(random.uniform(1000, 30000), 2)
    data_recebimento = fake.date_between(start_date='-2y', end_date='today')
    
    receitas.append([i, id_cliente, descricao, valor, data_recebimento])

df_receitas = pd.DataFrame(receitas, columns=["id_receita", "id_cliente", "descricao", "valor", "data_recebimento"])

# Salvando os arquivos em CSV
# df_clientes.to_csv("clientes.csv", index=False)
# df_faturas.to_csv("faturas.csv", index=False)
# df_pagamentos.to_csv("pagamentos.csv", index=False)
# df_despesas.to_csv("despesas.csv", index=False)
# df_receitas.to_csv("receitas.csv", index=False)
```
